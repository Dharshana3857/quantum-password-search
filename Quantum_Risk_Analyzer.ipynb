```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import math

def classical_complexity(n_bits):
    return 2 ** int(n_bits)

def quantum_complexity(n_bits):
    N = 2 ** int(n_bits)
    return int((math.pi / 4) * math.sqrt(N))

def effective_quantum_bits(n_bits):
    return n_bits / 2

def risk_level(n_bits):
    effective_bits = effective_quantum_bits(n_bits)

    if effective_bits < 40:
        return "CRITICAL"
    elif effective_bits < 64:
        return "HIGH"
    elif effective_bits < 96:
        return "MODERATE"
    else:
        return "LOW"

def analyze_security(n_bits):
    classical = classical_complexity(n_bits)
    quantum = quantum_complexity(n_bits)
    effective = effective_quantum_bits(n_bits)
    risk = risk_level(n_bits)

    print(" SECURITY ANALYSIS REPORT")
  
    print(f"Key Size: {n_bits} bits")
    print(f"Classical Worst-Case Attempts: {classical:e}")
    print(f"Quantum (Grover) Iterations: {quantum:e}")
    print(f"Effective Quantum Security: {effective} bits")
    print(f"Risk Level: {risk}")
```

```python
for bits in [4, 8, 16, 32, 64, 128,256]:
    analyze_security(bits)
    print("\n")
```
